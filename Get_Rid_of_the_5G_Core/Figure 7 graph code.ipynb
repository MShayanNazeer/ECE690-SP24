```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
import seaborn as sns
```

```python
idle_mode = np.loadtxt("data_for_figure_7.txt", usecols=0, dtype = float)
```

```python
N = len(idle_mode[600:1200])
ind = np.arange(N)


fig, ax = plt.subplots()
plt.figure(figsize=(4,1))

p1 = ax.plot(ind, idle_mode[600:1200], color = 'blue', linewidth = 1)

ax.set_xlabel('Time (s)', fontsize=20)
ax.set_ylabel('Power (mW)', fontsize=20)
ax.grid()

plt.show()
```
